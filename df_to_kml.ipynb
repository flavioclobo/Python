```python
with open("exportado.kml", "w") as f:
    # cabeçalho do kml
    f.writelines('<?xml version="1.0" encoding="UTF-8"?>\n')
    f.writelines('<kml xmlns="http://www.opengis.net/kml/2.2" xmlns:gx="http://www.google.com/kml/ext/2.2" xmlns:kml="http://www.opengis.net/kml/2.2" xmlns:atom="http://www.w3.org/2005/Atom">\n')
    f.writelines('<Document>\n')
    f.writelines('    <name>Dados do DataFrame exportado para kml</name>\n')
    f.writelines('    <Style id="style_1">\n') # definir um estilo para grupo de marcadores
    f.writelines('        <IconStyle>\n')
    f.writelines('            <Icon>\n')
    f.writelines('                <href>http://maps.google.com/mapfiles/kml/paddle/blu-circle.png</href>\n') # tipo do ícone
    f.writelines('            </Icon>\n')
    f.writelines('        </IconStyle>\n')
    f.writelines('    </Style>')
    f.writelines('    <Style id="style_2">\n') # segundo estilo
    f.writelines('        <IconStyle>\n')
    f.writelines('            <Icon>\n')
    f.writelines('                <href>http://maps.google.com/mapfiles/kml/paddle/red-circle.png</href>\n') # outro tipo de ícone
    f.writelines('            </Icon>\n')
    f.writelines('        </IconStyle>\n')
    f.writelines('    </Style>\n')
    f.writelines('    <Folder id="Pontos turísticos">\n')
    f.writelines('        <name>Nome do Folder</name>\n')

    # Cada ponto no mapa deverá ter a estrutura abaixo
    for index, row in df.iterrows():
        f.writelines('        <Placemark>\n')
        f.writelines('            <visibility>1</visibility>\n') # 0 oculto, 1 exibido
        f.writelines(f'            <name>{row['Descrição']}</name>\n')
        f.writelines(f'            <description>\n')
        
        if not pd.isna(row['Endereço']):
            f.writelines(f'                <strong>{row['Endereço']}</strong><br/><br/>\n')
                         
        if not pd.isna(row['Telefone']):
            f.writelines(f'                {row['Telefone']}<br/>\n')
        
        if not pd.isna(row['Website']):
            f.writelines(f'                {row['Website']}<br/><br/><a href="{row['Website']}"></a>\n')
        
        f.writelines(f'            </description>\n')
        f.writelines('            <styleUrl>#style_1</styleUrl>\n') # estilo do ponto
        f.writelines('            <Point>\n')
        f.writelines(f'                <coordinates>{row['Longitude']},{row['Latitude']}</coordinates>\n')
        f.writelines('            </Point>\n')
        f.writelines('        </Placemark>\n')

    f.writelines('    </Folder>\n')
    f.writelines('</Document>\n')
    f.writelines('</kml>\n')
```
