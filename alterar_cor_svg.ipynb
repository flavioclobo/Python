```python
from IPython.core.display import SVG
```

```python
file = input('Digite o nome do arquivo para ser alterado: ')
SVG(file)
```

```python
color_old = input('Digite o hexadecimal da cor para ser alterada: ')
color_new = input('Digite o hexadecimal da nova cor: ')
```

```python
with open(file) as f:
    contents = f.read()

contents = contents.replace(color_old, color_new)

with open(f'alterado_{file}', 'w') as f:
    for line in contents.split('\n'):
        f.write(f'{line}\n')
```
